> The below function extracts and adds my database + table into a Databricks Environment

```python
from mylib.extract import extract

# Define URL and table details
url = "https://raw.githubusercontent.com/nogibjj/chris_moreira_week6_sql_databricks/refs/heads/main/data/Spotify_Most_Streamed_Songs.csv"
database = "csm_87_database"
table_name = "csm_87_Spotify_Table"

# Run the extract function
extract(url, table_name, database)
```

The function below performs a transformation on the daata that was generated by extract

```python
from pyspark.sql import SparkSession
from mylib.transform import transform_data

# Initialize Spark session
spark = SparkSession.builder.appName("Transform Data").getOrCreate()

# Read data from Delta table
df = spark.table(f"{database}.{table_name}")

# Transform the data
transformed_df = transform_data(df)
transformed_df.show()
```

Lastly, we load this transformed data table into the databricks database

```python
from mylib.load import load_data

# Load the transformed data back into the Delta table
load_data(transformed_df, table_name, database)
```
